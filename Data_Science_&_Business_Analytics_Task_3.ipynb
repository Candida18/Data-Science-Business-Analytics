### **Distribution of Data**

```python
fig, axs = plt.subplots(ncols=2, nrows = 2, figsize = (10,10))
sns.distplot(df['Sales'], color = 'red',  ax = axs[0][0])
sns.distplot(df['Profit'], color = 'green',  ax = axs[0][1])
sns.distplot(df['Quantity'], color = 'orange',  ax = axs[1][0])
sns.distplot(df['Discount'], color = 'blue',  ax = axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize = 20)
axs[0][1].set_title('Profit Distribution', fontsize = 20)
axs[1][0].set_title('Quantity distribution', fontsize = 20)
axs[1][1].set_title('Discount Distribution', fontsize = 20)
plt.show()
```

### **Sub-categorywise Analysis of Sales, Discount and Profit**

```python
df_sub_category = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub_category.head(10)
```

_**Sub-categorywise Sales Analysis**_

```python
df_sub_category=df_sub_category.sort_values('Sales')

df_sub_category[['Sales']].plot(kind = 'bar', figsize = (20,5),color="#06327d")
plt.title('Sub-categorywise Sales Analysis', fontsize = 20)
plt.ylabel('Sales per Sub-category')
plt.xlabel('Sub-category')
plt.show()
```

_**Sub-categorywise Discount Analysis**_

```python
df_sub_category=df_sub_category.sort_values('Discount')

df_sub_category[['Discount']].plot(kind = 'bar', figsize = (20,5),color="#06327d")
plt.title('Sub-categorywise Discount Analysis', fontsize = 20)
plt.ylabel('Discount per Sub-category')
plt.xlabel('Sub-category')
plt.show()
```

_**Sub-categorywise Profit Analysis**_

```python
df_sub_category.sort_values('Profit')[['Sales','Profit']].plot(kind='bar',
                                                              figsize= (20,5),
                                                              label=['Avg Sales Price($)','Profit($)'],color=['#050a30','#5cb6f9'])
```
